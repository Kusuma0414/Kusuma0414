```python
import pandas as pd
df=pd.read_excel('/content/clasic.xlsx')
numeric_df=df.select_dtypes(include=['number'])
z_score=(numeric_df-numeric_df.mean())/numeric_df.std()
outliers=df[abs(z_score)> -1]
print(outliers)
```

```python
import pandas as pd
df=pd.read_excel('/content/clasic.xlsx')
numeric_data=df.select_dtypes(include=['number'])
rolling_mean=numeric_data.rolling(window=5).mean()
rolling_std=numeric_data.rolling(window=5).std()
outliers=numeric_data[(numeric_data-rolling_mean).abs() >-1* rolling_std]
print(outliers)
```
