```python
import pandas as pd
d = pd.read_excel('/content/lsg_stats_2024_q2.xlsx')
d['Sales 2Q 2023'] = d['Sales 2Q 2023'].map(lambda x: x * 1.10)
print(d)
```

```python
filtered_d = d[d['Median 2Q 2023'].map(lambda x: x > 1000)]
print(filtered_d)
```

```python
from functools import reduce
if not filtered_d['Median 2Q 2023'].empty:
  total=reduce(lambda x, y: x + y, filtered_d['Median 2Q 2023'])
  print(total)
else:
  pass
  print("the filtered DataFrame is empty,cannot calculate total Median 2Q 2023")
```

```python
import pandas as pd
d=pd.read_excel("/content/lsg_stats_2024_q2.xlsx")
import matplotlib.pyplot as plt
plt.plot(d['Median 2Q 2023'], d['Median Change'], marker='o',linestyle='-',color='red')
plt.title("line plot")
plt.xlabel("Median 2Q 2023")
plt.ylabel("Median Change")
plt.show()
```
